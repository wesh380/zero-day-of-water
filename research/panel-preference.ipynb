# Panel Preference Model
این نوت‌بوک داده‌های استفاده از پنل‌ها را بارگذاری کرده و یک مدل رگرسیون لجستیک ساده برای پیش‌بینی احتمال کلیک روی هر پنل می‌سازد.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

# داده‌ها را بارگذاری کنید
df = pd.read_csv('../data/panel-usage.csv')
X = pd.get_dummies(df['panel'], prefix='panel', drop_first=True)
y = df['clicked']

# مدل رگرسیون لجستیک
model = LogisticRegression()
model.fit(X, y)

coef = model.coef_[0]
intercept = model.intercept_[0]
probs = {
    'A': 1 / (1 + np.exp(-(intercept))),
    'B': 1 / (1 + np.exp(-(intercept + coef[0]))),
    'C': 1 / (1 + np.exp(-(intercept + coef[1])))
}
print('Predicted probabilities:', probs)
print('Recommended order:', sorted(probs, key=probs.get, reverse=True))
```
